```python
import os
import glob
import numpy as np
import pandas as pd
from nltools.data import Brain_Data
import nibabel as nib
```

```python
data_dir = '/mnt/chrastil/lab/users/lily/path_direction/localizer_beta_series_all_run_exploration'
```

```python
sub_list = glob.glob(os.path.join(data_dir, 'sub*'))
```

```python
print(sub_list)
```

```python
colnames = ['sub', 'exp', 'exp2', 'test1', 'test2', 'test3', 'test4', 'test5', 'test6']
scan_time = pd.DataFrame(np.zeros((len(sub_list), len(colnames))), columns = colnames)
counter = 0
for i in sub_list:
    sub = str.split(i, '/')[-1]
    
    input_subnum = str.split(sub,'-')[-1]
    scan_time.iloc[counter]['sub'] = int(input_subnum) # store subject number
    
    exp_name = glob.glob(os.path.join(data_dir, sub, 'func', '*boldEx_*.nii.gz'))
    exp = ''.join([str(nib.load(file).shape[3]) for file in exp_name])
    scan_time.iloc[counter]['exp'] = exp # store exploration 1
    
    exp2_name = glob.glob(os.path.join(data_dir, sub, 'func', '*boldEx2_*.nii.gz'))
    exp2 = ''.join([str(nib.load(file).shape[3]) for file in exp2_name])
    scan_time.iloc[counter]['exp2'] = exp2 # store exploration 2
    
    test1_name = glob.glob(os.path.join(data_dir, sub, 'func', '*boldRun1_*.nii.gz'))
    test1 = ''.join([str(nib.load(file).shape[3]) for file in test1_name])
    scan_time.iloc[counter]['test1'] = test1 # store test1
    
    test2_name = glob.glob(os.path.join(data_dir, sub, 'func', '*boldRun2_*.nii.gz'))
    test2 = ''.join([str(nib.load(file).shape[3]) for file in test2_name])
    scan_time.iloc[counter]['test2'] = test2 # store test2
    
    test3_name = glob.glob(os.path.join(data_dir, sub, 'func', '*boldRun3_*.nii.gz'))
    test3 = ''.join([str(nib.load(file).shape[3]) for file in test3_name])
    scan_time.iloc[counter]['test3'] = test3 # store test3
    
    test4_name = glob.glob(os.path.join(data_dir, sub, 'func', '*boldRun4_*.nii.gz'))
    test4 = ''.join([str(nib.load(file).shape[3]) for file in test4_name])
    scan_time.iloc[counter]['test4'] = test4 # store test4
    
    test5_name = glob.glob(os.path.join(data_dir, sub, 'func', '*boldRun5_*.nii.gz'))
    test5 = ''.join([str(nib.load(file).shape[3]) for file in test5_name])
    scan_time.iloc[counter]['test5'] = test5 # store test5
    
    test6_name = glob.glob(os.path.join(data_dir, sub, 'func', '*boldRun6_*.nii.gz'))
    test6 = ''.join([str(nib.load(file).shape[3]) for file in test6_name])
    scan_time.iloc[counter]['test6'] = test6 # store test6
    
    
    counter += 1
    
# save data file
str_name = data_dir + '/scan_time.csv'
scan_time.to_csv(str_name, index=False)
```
